```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
top_10_locations = airplane['Location'].value_counts().nlargest(10).index
top_10_operator = airplane['Operator'].value_counts().nlargest(10).index

airplane_top_10_locations_operators = airplane[
    airplane['Location'].isin(top_10_locations) &
    airplane['Operator'].isin(top_10_operator)
]

sns.relplot(
    data = airplane_top_10_locations_operators,
    x = "Location" , y ="Operator"
)
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
airplane['Year'] = pd.to_datetime(airplane['Date']).dt.year
accidents_per_year = airplane['Year'].value_counts().sort_index()

plt.figure(figsize=(12, 6))
sns.lineplot(x=accidents_per_year.index, y=accidents_per_year.values)
plt.xlabel('Year')
plt.ylabel('Number of Accidents')
plt.title('Number of Airplane Accidents Over Time')
plt.grid(True)
plt.show()
```

```python
fatalities_per_year = airplane.groupby('Year')['Fatalities'].sum()
aboard_per_year = airplane.groupby('Year')['Aboard'].sum()
fatality_rate_per_year = (fatalities_per_year / aboard_per_year).fillna(0) * 100

plt.figure(figsize=(12, 6))
sns.lineplot(x=fatality_rate_per_year.index, y=fatality_rate_per_year.values)
plt.xlabel('Year')
plt.ylabel('Fatality Rate (%)')
plt.title('Airplane Accident Fatality Rate Over Time')
plt.grid(True)
plt.show()
```

```python
numerical_cols = airplane.select_dtypes(include=np.number).columns
correlation_matrix = airplane[numerical_cols].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=airplane, x='Aboard', y='Fatalities')
plt.xlabel('Aboard')
plt.ylabel('Fatalities')
plt.title('Relationship between Aboard and Fatalities')
plt.show()
```
